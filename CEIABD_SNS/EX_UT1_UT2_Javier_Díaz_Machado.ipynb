3) Estamos analizando un conjunto de datos aleatorios y deseamos identificar los outliers. 
Para lo cual, es necesario generar 50 números aleatorios con una distribución normal, con media 50 y desviación estándar 15. La variable donde se guarden se ha de llamar: datosEstudio
Y en la posición 10, 20 y 30 guardar los valores 105, 200 y 300 respectivamente.
Para identificar los outliers, se debe utilizar los siguientes métodos:
    a) Utilizando la técnica basada en probabilidad
    b) Utilizando la técnica basada en los cuartiles
        i) Para los outliers leves (mild outlier): si está fuera del rango:  [ Q1 - 1.5 x | QR, Q3 + 1.5 x | QR]
        ii) Para los outliers extremos (extreme outlier): si está fuera del rango:  [ Q1 - 3 x | QR, Q3 + 3 x | QR]
    c) ¿Qué diferencias observas entre los resultados obtenidos con ambas técnicas (probabilidad y quartiles)? Y argumenta tu respuesta.




la función np.abs(datosEstudio-valorMedia).

```python
datosEstudio = np.random.normal(50, 15, 50)
datosEstudio[10], datosEstudio[20], datosEstudio[30] = 105, 200, 300

#a) métrica basada en probabilidad:
p_g=0.85   
# probabilidad global
alfa_g=(1-p_g)/2
# probabilidad  para un solo dato
alfa= 1-(1-alfa_g)**(1/len(datosEstudio)) 
Z_alfa=st.norm.ppf(1-alfa/2)
print(f" Alfa ={alfa}")
print(f" Z_alfa ={Z_alfa}")

xL=  round(np.mean(datosEstudio)-Z_alfa*  np.std(datosEstudio),4)
xU=  round(np.mean(datosEstudio)+Z_alfa*  np.std(datosEstudio),4)
print(f" Banda= [ {xL},{xU}]")
for i in  range(len(datosEstudio)):
        if datosEstudio[i] < xL  or datosEstudio[i]>xU:
          print(f" El dato[{i}]={datosEstudio[i]} es un outlier")
```

```python
#    b) Utilizando la técnica basada en los cuartiles
#        i) Para los outliers leves (mild outlier): si está fuera del rango:  [ Q1 - 1.5 x | QR, Q3 + 1.5 x | QR]
#        ii) Para los outliers extremos (extreme outlier): si está fuera del rango:  [ Q1 - 3 x | QR, Q3 + 3 x | QR]
Q1 = np.quantile(datosEstudio,0.25)
Q3 = np.quantile(datosEstudio,0.75)
IQR = Q3 - Q1 # Ver la gráfica para entender el concepto de IQR (Rango intercuartílico)
#k=1.5 # k=1,5 es un outlier leve y k=3 es un outlier extremo
k=1.8 # Variando este valor lograremos hacer las bandas más o menos estrechas ó anchas
xL=Q1 - k * IQR
xU=Q3 + k * IQR
print(f" Banda= [ {xL},{xU}]")
for i in  range(len(datosEstudio)):
        if datosEstudio[i] < xL  or datosEstudio[i]>xU:
          print(f" El dato[{i}]={datosEstudio[i]} es un outlier")
```
