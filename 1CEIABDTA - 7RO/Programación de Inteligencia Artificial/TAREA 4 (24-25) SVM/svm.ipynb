```
#Importamos las librerías
import pandas as pd
# Descargamos e importamos el dataset
import kagglehub
path = kagglehub.dataset_download("uciml/iris")
df = pd.read_csv(path+'/Iris.csv')
```

```
from sklearn.model_selection import train_test_split
#Definimos los datos de entrenamiento y prueba

training, test = train_test_split(df, test_size=0.2)
trainX = training.drop(['Id','Species'], axis=1)/10
trainY = training['Species']
testX = test.drop(['Id','Species'], axis=1)/10
testY = test['Species']
```

```
#Podemos ver los datos:
print("  Training Data:\n",trainY.head(),'\n', trainX.head(), '\n\n  Test Data:\n', testY.head(), '\n', testX.head())
```

```
from sklearn.svm import SVC
# Entrenamos al modelo con los datos
svc = SVC()
predict = svc.fit(trainX, trainY)

df = pd.DataFrame({
    'Original specie': testY,
    'Predicted specie': predict.predict(testX),
    'Success': predict.predict(testX) == testY
    }
)
df
```
